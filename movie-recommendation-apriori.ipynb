```python
import numpy as np 
import pandas as pd 
from apyori import apriori

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
movie_titles_df = pd.read_csv('../input/dataset/Movie_Id_Titles.csv')
movie_rating_df = pd.read_csv('../input/dataset/Dataset.csv')
```

```python
movie_rating_df.drop('timestamp',axis=1,inplace=True)
movie_rating_df.head()
```

```python
main_df = pd.merge(movie_rating_df,movie_titles_df,on='item_id')
main_df
```

```python
main_df.drop(['item_id'],axis=1,inplace=True)
main_df
```

```python
main_df = main_df.pivot_table(index='user_id',columns='title',values='rating')
main_df
```

```python
main_df.fillna(0,inplace=True)
main_df
```

```python
movie_list = []
for i in range(0,944):
    movie_list.append([str(main_df.columns[j]) for j in range(0,1664) if main_df.values[i,j] != 0 ])
```

```python
rules = apriori(transactions = movie_list,min_support=0.05,min_confidence=0.2,min_lift=6,max_length=2)
results = list(rules)
```

```python
def data_extractor(data):
    m1 = np.array([tuple(i[2][0][0] for i in data)]).T
    m1 = pd.DataFrame(m1,columns=['Movie_Watched'])
    m2 = np.array([tuple(i[2][0][1] for i in data)]).T
    m2 = pd.DataFrame(m2,columns=['Also_Watched'])
    support = pd.DataFrame([i[1] for i in data],columns=['Support'])
    confidence = pd.DataFrame([i[2][0][2] for i in data],columns=['Confidence'])
    lift = pd.DataFrame([i[2][0][3] for i in data],columns=['Lift'])
    result = pd.concat([m1,m2,support,confidence,lift],axis=1)
    return result
result_df = data_extractor(results)
```

```python
result_df.sort_values(by='Lift',ascending=False)
```
